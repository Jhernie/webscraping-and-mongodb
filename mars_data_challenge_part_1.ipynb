```python
# import dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Navigate to the landing page for scraping
url = 'https://redplanetscience.com/'

browser.visit(url)
```

```python
# Get the HTML content
html = browser.html
```

```python
# Use Beautiful Soup to parse the HTML content
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
# Find all the articles on the landing page
articles = soup.find_all('div', class_='list_text')
```

```python
# Loop through returned queries
for article in articles:

    # Retrieve the article title
    title = article.find('div', class_='content_title').text.strip()

    # Retrieve the article date
    date = article.find('div', class_="list_date").text
    
    data = {
        "TITLE": title,
        "DATE" : date
    }

 
    print(data)
    print('-----------------')
```

```python
# Close the browser
browser.quit()
```
